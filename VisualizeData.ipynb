Visualizing Student Preformance Factors Dataset

Goal is to see if we can visually identify any potentially correlated features

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV data
df = pd.read_csv("StudentPerformanceFactors.csv")

target_col = "Exam_Score"

# Identify numeric and categorical columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object", "bool", "category"]).columns

# Exclude the target itself from the numeric columns (in case it's numeric)
if target_col in numeric_cols:
    numeric_cols = numeric_cols.drop(target_col)

# 1) Plot scatter plots for numeric features vs. Exam_Score
for col in numeric_cols:
    plt.figure()
    plt.scatter(df[col], df[target_col])
    plt.title(f"{target_col} vs. {col}")
    plt.xlabel(col)
    plt.ylabel(target_col)
    plt.show()

# 2) Plot box plots for categorical features vs. Exam_Score
for col in categorical_cols:
    # Some columns might have many unique categories, so be mindful of that
    plt.figure()
    df.boxplot(column=target_col, by=col)
    plt.title(f"{target_col} by {col}")
    plt.suptitle("")  # Remove the default "Boxplot grouped by" subtitle
    plt.xlabel(col)
    plt.ylabel(target_col)
    plt.show()
```
